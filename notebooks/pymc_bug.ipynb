```python
import pymc3 as pm
import numpy as np

data = np.random.poisson(lam=10, size=200)

model = pm.Model()
with model:
    mu = pm.HalfFlat('sigma')
    pm.Normal('foo', mu=mu, observed=data)
    
    trace = pm.sample(tune=1000)

with model:
    samples = pm.sample_ppc(trace)
```

```python
import pymc3 as pm
import numpy as np

data = np.random.normal(loc=4, size=500)

model = pm.Model()
with model:
    sd = pm.HalfFlat('sigma')
    pm.Normal('foo', mu=4, sd=sd, observed=data)
    
    trace = pm.sample()

with model:
    samples = pm.sample_ppc(trace)
```
